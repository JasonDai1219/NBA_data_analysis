```python
import pandas as pd
import numpy as np
import requests
import time
import bs4
```

# Web Scraping

```python
salary_response = requests.get('https://www.espn.com/nba/salaries/_/year/2023/page/1/seasontype/4')
salary_response
```

```python
soup = bs4.BeautifulSoup(salary_response.text)
small_total_page = int(soup.find('div', attrs = {'class' : 'page-numbers'}).text.split(' ')[-1])
small_total_page
```

```python
seasons = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
specific_seasons = ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23']
```

```python
dfs_list = []
counter = 0
for s in seasons:
    season_salary_response = requests.get(f'https://www.espn.com/nba/salaries/_/year/{s}/seasontype/4')
    small_total_page = int(soup.find('div', attrs = {'class' : 'page-numbers'}).text.split(' ')[-1])
    names = []
    salaries = []
    for i in range(1, small_total_page + 1):
        salary_response = requests.get(f'https://www.espn.com/nba/salaries/_/year/{s}/page/{i}/seasontype/4')
        soup = bs4.BeautifulSoup(salary_response.text)
        players_info = soup.find('table', attrs = {'class' : 'tablehead'}).find_all('tr')[1:]
        for j in players_info:
            if (j.get('class') == ['colhead']):
                continue
            names.append(j.find('a').text)
            salaries.append(j.find('td', attrs = {'style' : 'text-align:right;'}).text)
        lag = np.random.uniform(low = 5, high = 10)
        time.sleep(lag)
    years = [specific_seasons[counter] for i in range(len(names))]
    counter += 1
    df_dict = {'Name' : names, 'Salary' : salaries, 'Season' : years}
    df = pd.DataFrame(df_dict)
    dfs_list.append(df)
```

```python
all_salaries = pd.concat(dfs_list, axis = 0)
all_salaries.to_csv('/Users/jasondai/Desktop/NBA_salary_data.csv', index=False)
all_salaries
```

```python
new_headers = ['Year', 'Season_Type'] + the_columns
```

```python
the_df = pd.DataFrame(columns = new_headers)
Season_Types = ['Regular%20Season', 'Playoffs']
Years = ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23']

start_loop = time.time()
for y in Years:
    for s in Season_Types:
        the_url = f" https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season={y}&SeasonType={s}&StatCategory=PTS"
        r = requests.get(the_url).json()
        the_columns = r['resultSet']['headers']
        temp1 = pd.DataFrame(r['resultSet']['rowSet'], columns = the_columns)
        Year = {'Year' : [y for i in range(len(temp1))], 'Season_Type' : [s for i in range(len(temp1))]}
        temp2 = pd.concat([pd.DataFrame(Year), temp1], axis = 1)
        the_df = pd.concat([the_df, temp2], axis = 0)
        print(f'accomplished{y}{s}.')
        lag = np.random.uniform(low = 5, high = 40)
        time.sleep(lag)
the_df
print('Total Time to accomplish the process is' + str(time.time() - start_loop))
```
